```python
df4 = pd.read_excel('2020_Europeans_firstrounds.xlsx')
df4.rename(columns={"Unnamed: 2": "Fighter_1_elo", "Unnamed: 3": "Win_expec",
                   "Unnamed: 7": "Fighter_2_elo", "Unnamed: 8": "Rating_Dif",
                   "Unnamed: 9": "Win_expec_2"}, inplace=True)
```

```python
fighter_dict = dict(zip(feather.fighter, feather.rating))
df4['Fighter_1_elo'] = df4['Fighter'].map(fighter_dict)
df4['Fighter_2_elo'] = df4['Opponent'].map(fighter_dict)
```

```python
# mapping all weight classes
eu_rooster = df4.loc[df4.Weight == 'Rooster']
eu_light_feather = df4.loc[df4.Weight == 'Light Feather']
eu_feather = df4.loc[df4.Weight == 'Feather']
eu_light = df4.loc[df4.Weight == 'Light']
eu_middle = df4.loc[df4.Weight == 'Middle']
eu_medium_heavy = df4.loc[df4.Weight == 'Medium Heavy']
eu_heavy = df4.loc[df4.Weight == 'Heavy']
eu_super_heavy = df4.loc[df4.Weight == 'Super Heavy']
eu_ultra_heavy = df4.loc[df4.Weight == 'Ultra Heavy']

fighter_dict7 = dict(zip(super_heavy.fighter, super_heavy.rating))
eu_super_heavy['Fighter_1_elo'] = eu_super_heavy['Fighter'].map(fighter_dict7)
eu_super_heavy['Fighter_2_elo'] = eu_super_heavy['Opponent'].map(fighter_dict7)

fighter_dict7 = dict(zip(ultra_heavy.fighter, ultra_heavy.rating))
eu_ultra_heavy['Fighter_1_elo'] = eu_ultra_heavy['Fighter'].map(fighter_dict7)
eu_ultra_heavy['Fighter_2_elo'] = eu_ultra_heavy['Opponent'].map(fighter_dict7)

fighter_dict1 = dict(zip(rooster.fighter, rooster.rating))
eu_rooster['Fighter_1_elo'] = eu_rooster['Fighter'].map(fighter_dict1)
eu_rooster['Fighter_2_elo'] = eu_rooster['Opponent'].map(fighter_dict1)

fighter_dict2 = dict(zip(light_feather.fighter, light_feather.rating))
eu_light_feather['Fighter_1_elo'] = eu_light_feather['Fighter'].map(fighter_dict2)
eu_light_feather['Fighter_2_elo'] = eu_light_feather['Opponent'].map(fighter_dict2)

fighter_dict3 = dict(zip(feather.fighter, feather.rating))
eu_feather['Fighter_1_elo'] = eu_feather['Fighter'].map(fighter_dict3)
eu_feather['Fighter_2_elo'] = eu_feather['Opponent'].map(fighter_dict3)

fighter_dict4 = dict(zip(light.fighter, light.rating))
eu_light['Fighter_1_elo'] = eu_light['Fighter'].map(fighter_dict4)
eu_light['Fighter_2_elo'] = eu_light['Opponent'].map(fighter_dict4)

fighter_dict5 = dict(zip(middle.fighter, middle.rating))
eu_middle['Fighter_1_elo'] = eu_middle['Fighter'].map(fighter_dict5)
eu_middle['Fighter_2_elo'] = eu_middle['Opponent'].map(fighter_dict5)

fighter_dict6 = dict(zip(medium_heavy.fighter, medium_heavy.rating))
eu_medium_heavy['Fighter_1_elo'] = eu_medium_heavy['Fighter'].map(fighter_dict6)
eu_medium_heavy['Fighter_2_elo'] = eu_medium_heavy['Opponent'].map(fighter_dict6)

fighter_dict7 = dict(zip(heavy.fighter, heavy.rating))
eu_heavy['Fighter_1_elo'] = eu_heavy['Fighter'].map(fighter_dict7)
eu_heavy['Fighter_2_elo'] = eu_heavy['Opponent'].map(fighter_dict7)

df5 = pd.concat([eu_rooster,eu_light_feather,eu_feather,eu_light,eu_middle,
                eu_medium_heavy,eu_heavy,eu_super_heavy,eu_ultra_heavy])
df5['Rating_Dif'] = df5.Fighter_1_elo - df5.Fighter_2_elo
df5.head(50)
```

```python
import  math
def Probability_1(rating1, rating2):
    return 1.0 * 1.0 / (1 + 1.0 * math.pow(10, -1.0 * (rating1 - rating2) / 400))
def Probability_2(rating1, rating2):
    return 1.0 * 1.0 / (1 + 1.0 * math.pow(10, -1.0 * (rating2 - rating1) / 400))
```

```python
df5['Win_expec'] = df5.apply(lambda x: Probability_1(x.Fighter_1_elo,x.Fighter_2_elo),axis=1)
df5['Win_expec_2'] = df5.apply(lambda x: Probability_2(x.Fighter_1_elo,x.Fighter_2_elo),axis=1)
df5['Rating_Dif'] = df5.Fighter_1_elo - df5.Fighter_2_elo
display(df5.head(50))
display(df5.tail(30))
```
