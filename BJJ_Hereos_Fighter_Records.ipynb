# Scrape BJJ Hereos for Fighter Data

#### Step 1: Get list of fighters' page urls
- https://www.bjjheroes.com/a-z-bjj-fighters-list

```python
import re
import requests
from bs4 import BeautifulSoup
import sys
import csv
import urllib
import time
```

```python
html_page = requests.get('https://www.bjjheroes.com/a-z-bjj-fighters-list')
soup_main = BeautifulSoup(html_page.text, 'html.parser')
```

```python
fighter_list = soup_main.find('tbody', {'class':'row-hover'})

for link in fighter_list.find_all('a'): 
    print("https://www.bjjheroes.com" + link.get('href') )
```

```python
for link in fighter_list.find_all('a'): 
    display("https://www.bjjheroes.com" + link.get('href') )
```

```python
fighter_list = soup_main.find('tbody', {'class':'row-hover'})

fighters_urls = []

for link in fighter_list.find_all('a'):

#     print("https://www.bjjheroes.com" + link.get('href') )
    url_name = "https://www.bjjheroes.com" + link.get('href')
    fighters_urls.append(url_name)
    
df = pd.DataFrame(fighters_urls, columns=['urls'])
df.drop_duplicates(inplace=True)
print(df.shape)
df.head(3)
```

```python
final_urls_list = df.urls.tolist()
```

#### Step 2: Download Fight History
- https://www.bjjheroes.com/bjj-fighters/adam-wardzinski

Amedned Fighters List with Fight History

```python
results = []
for url in final_urls_list:
    try:
        time.sleep(5)
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')

        fight_record = soup.find('h3', {'class': 'heading'}).get_text(strip=True)
            
        results.append((fight_record))
            
        print ('Processed url: %s' %url)
    except:
        print ('Skipped - No fight history: %s' %url)
        
df1 = pd.DataFrame(pd.DataFrame(results, columns=['fight_record',]))
```

```python
display(df1.head(5))
display(df1.tail(5))
display(df1.shape)
df_list = df1.fight_record.tolist()
pd.DataFrame(df_list).to_excel('Fighters_Fight_History.xlsx', header=False, index=False)
```
